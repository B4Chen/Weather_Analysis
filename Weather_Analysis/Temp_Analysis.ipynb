```python
import pandas as pd


df = pd.read_csv("/Users/brian/Pandas/cityTemp.csv")
df
```

```python
### Get the cities only in United State
df_us = df.loc[df['Country'] == 'United States']

# Print the filtered DataFrame
print(df_us)
```

```python
### Get the Farenheight of each month

df = df[df['Country'] == 'United States'].copy()
# Extract the values within parentheses for each month fahrenheit 
df['Jan_Value'] = df['Jan'].str.split('\n').str[1].str.strip('()')
df['Feb_Value'] = df['Feb'].str.split('\n').str[1].str.strip('()')
df['Mar_Value'] = df['Mar'].str.split('\n').str[1].str.strip('()')
df['Apr_Value'] = df['Apr'].str.split('\n').str[1].str.strip('()')
df['May_Value'] = df['May'].str.split('\n').str[1].str.strip('()')
df['Jun_Value'] = df['Jun'].str.split('\n').str[1].str.strip('()')
df['Jul_Value'] = df['Jul'].str.split('\n').str[1].str.strip('()')
df['Aug_Value'] = df['Aug'].str.split('\n').str[1].str.strip('()')
df['Sep_Value'] = df['Sep'].str.split('\n').str[1].str.strip('()')
df['Oct_Value'] = df['Oct'].str.split('\n').str[1].str.strip('()')
df['Nov_Value'] = df['Nov'].str.split('\n').str[1].str.strip('()')
df['Dec_Value'] = df['Dec'].str.split('\n').str[1].str.strip('()')
# Display the extracted values along with the city and month
result_df = df[['City', 'Jan_Value','Feb_Value','Mar_Value', 'Apr_Value', 'May_Value', 'Jun_Value', 'Jul_Value',
                'Aug_Value','Sep_Value','Nov_Value','Dec_Value',]]
print(result_df)
```

```python
### Get the Celius of each month

df_us = df[df['Country'] == 'United States'].copy()

df_us['Jan_Value'] = df_us['Jan'].str.split('\n').str[0]
df_us['Feb_Value'] = df_us['Feb'].str.split('\n').str[0]
df_us['Mar_Value'] = df_us['Mar'].str.split('\n').str[0]
df_us['Apr_Value'] = df_us['Apr'].str.split('\n').str[0]
df_us['May_Value'] = df_us['May'].str.split('\n').str[0]
df_us['Jun_Value'] = df_us['Jun'].str.split('\n').str[0]
df_us['Jul_Value'] = df_us['Jul'].str.split('\n').str[0]
df_us['Aug_Value'] = df_us['Aug'].str.split('\n').str[0]
df_us['Sep_Value'] = df_us['Sep'].str.split('\n').str[0]
df_us['Oct_Value'] = df_us['Oct'].str.split('\n').str[0]
df_us['Nov_Value'] = df_us['Nov'].str.split('\n').str[0]
df_us['Dec_Value'] = df_us['Dec'].str.split('\n').str[0]
print(df_us[['Country', 'City', 'Jan_Value', 'Feb_Value', 'Mar_Value', 'Apr_Value', 'May_Value', 
             'Jun_Value', 'Jul_Value','Aug_Value', 'Sep_Value', 'Oct_Value', 'Nov_Value', 'Dec_Value']])
```

```python
### Get the city in United State where in the month of Jan the farenheight is greater than 30

df = df[df['Country'] == 'United States'].copy()
df['Jan Fahrenheit'] = df['Jan'].str.split('\n').str[1].str.strip('()').str.replace('−', '-').astype(float)
df_us = df[['City', 'Jan Fahrenheit']]
df_us = df_us[df_us['Jan Fahrenheit'] > 30]
df_us = df_us.rename(columns={'Jan Fahrenheit': 'Jan Fahrenheit'})

print(df_us)


```

```python
### Get the highest temperature in farenheight for each month and display the city
numeric_columns = ['Jan_Value', 'Feb_Value', 'Mar_Value', 'Apr_Value', 'May_Value', 'Jun_Value', 'Jul_Value', 'Aug_Value', 'Sep_Value', 'Oct_Value', 'Nov_Value', 'Dec_Value']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

highest_value_by_month = df[['City'] + numeric_columns]
highest_value_by_month.set_index('City', inplace=True)

for month in numeric_columns:
    city_with_highest_value = highest_value_by_month[month].idxmax()
    highest_value = highest_value_by_month.loc[city_with_highest_value, month]
    print("Month: {}, City: {}, Highest Value: {}".format(month, city_with_highest_value, highest_value))
```

```python
### Get the highest temperature in celcius for each month and display the city
numeric_columns = ['Jan_Value', 'Feb_Value', 'Mar_Value', 'Apr_Value', 'May_Value', 'Jun_Value', 'Jul_Value', 'Aug_Value', 'Sep_Value', 'Oct_Value', 'Nov_Value', 'Dec_Value']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

lowest_value_by_month = df[['City'] + numeric_columns]
lowest_value_by_month.set_index('City', inplace=True)

for month in numeric_columns:
    city_with_lowest_value = lowest_value_by_month[month].idxmin()
    lowest_value = lowest_value_by_month.loc[city_with_lowest_value, month]
    print("Month: {}, City: {}, Lowest Value: {}".format(month, city_with_lowest_value, lowest_value))
```
